```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import os
from tqdm.notebook import tqdm

%matplotlib inline
```

```python
root_path = '/media/alexey/HDDData/datasets/railway/RZD_Segmentation/train/'

basename = 'img_0.9021612708640853'

image_path = root_path + 'images/{}.png'.format(basename)
mask_path = root_path + 'mask/{}.png'.format(basename)
```

```python
image = cv2.cvtColor(cv2.imread(image_path, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)
mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
```

```python
print(np.unique(mask))

# 0 - bg
# 6 - second road
# 7 - main road
# 10 - other train

plt.figure(figsize=(12, 10))
plt.imshow(mask == 10)
plt.show()
```

```python
masks_folder = os.path.join(root_path, 'mask/')

labels = []

for mask_name in tqdm(os.listdir(masks_folder)):
    mask_path = os.path.join(masks_folder, mask_name)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    u_values = np.unique(mask)
    for v in u_values:
        if v not in labels:
            labels.append(v)
    
labels.sort() 
print(labels)
```
